### (2) 데이터 표준화

- 사용자 함수 정의 : age()

```python
# 2019년의 연령값은 20대~60대로 그룹화되어있어 2017,2021의 개별 연령값을 같은 형태로 변경

def age(n) :
    if (n//10) >= 6 :
        return 5
    if (n//10) == 5 :
        return 4
    if (n//10) == 4 :
        return 3
    if (n//10) == 3 :
        return 2
    else :
        return 1
```
