- 사용자 함수 정의 : age()

```python
# 2019년의 연령값은 20대~60대로 그룹화되어있어 2017,2021의 개별 연령값을 같은 형태로 변경

def age(n) :
    if (n//10) >= 6 :
        return 5
    if (n//10) == 5 :
        return 4
    if (n//10) == 4 :
        return 3
    if (n//10) == 3 :
        return 2
    else :
        return 1
```

## 3) 주이용 도서관 종류 빈도분석

- 필요한 컬럼 추출

```python
main_lib = total_stu.loc[:,['연도', '주이용 도서관 종류1', '주이용 도서관 종류2']]
```

```python
main_lib
```

- 연도기준 그룹화 : groupby()

```python
main_lib_gr = main_lib.groupby('연도')
```

- 함수 정의 : 주이용 도서관 1순위에 가중치 2를 부여, 합산한 값 구하기.
